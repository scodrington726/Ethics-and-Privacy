## Example of Misuse of Health Data

An example of misuse of health data is the case of Henrietta Lacks. Henrietta Lacks' cancer cells were taken without her knownledge nor consent during a cervical cancer biopsy at Johns Hopkins. Her cells that divided infinitely have contributed to the developing of medicine including the polio vaccine, cancer research, HIV, and Covid-19 studies. However, with this collection of her cells was highly unethical. First, they were collected without Lacks' informed consent. Second, her and her family for years were decived because of the false pseudonym her cells were held under. To prevent this from happening the providers could have properly informed and gained permission from Lacks to use her cells in this research.

## Example of a Breach in Health Data

An example of a breach in health data is the Yale New Haven Health case in 2025. This case invovled the unauthorized third-party access to private information including names, dates of birth, social security numbers, and medical records. This affected over 5.5 million people. Yale New Haven Health had to file for a 18 million settlement claim for those affected. Additionally, they had to notify all the individuals affected. Something that could have prevented this is extra security permissions and third party risk management.

## Bias in Health Data

An example of bias in health data collection is the temporal thermometers or also known as forehead thermometers. Forehead thermometers show racial bias in data collection. These thermometers tend to underestimate fevers in Black patients by about 26% compared to oral ones. This can lead to misdiagnoses and treatment delays. This can be prevented by collecting testing data on more diverse populations and ensuring consistency.

## Anoymization of Health Data

A method to anoymize health data is pseudonymization. This technique replaces personal data with made-up identifiers, codes or pseudonyms, to make data less linkable to individuals. This method still allows for re-identification. It ensures privacy for research by protecting the identities of those invovled.
